```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = []
with open("/Users/yuvalkirstain/school/repos/coref-transformers/one-epoch/test_to_write.jsonl") as f:
    for line in f:
        data.append(json.loads(line))
df = pd.DataFrame(data)
```

```python
df["step"] = df.data.apply(lambda x: int(x.split('_')[-1]))
```

```python
for col in df.columns:
    if "step" in col or "experiment" in col or "data" in col:
        continue
    df.plot(x="step", y=col, kind="line", title=f"{col} vs step")
    plt.savefig(f'/Users/yuvalkirstain/school/repos/coref-transformers/{col}_vs_step.png')
```

```python
df.plot(x="step", y="f1", kind="line", title="f1 vs step")
plt.savefig('/Users/yuvalkirstain/school/repos/coref-transformers/f1_vs_step.png')
```
